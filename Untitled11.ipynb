```python
import pandas as pd

# Load the Excel file
file_path = r'C:\Users\gaura\Downloads\Data Analyst Intern Assignment - Excel.xlsx'

# Check sheet names first to understand the structure
excel_data = pd.ExcelFile(file_path)
sheet_names = excel_data.sheet_names
sheet_names
```

```python
# Load each sheet into a DataFrame
user_details = pd.read_excel(file_path, sheet_name='UserDetails.csv')
cooking_sessions = pd.read_excel(file_path, sheet_name='CookingSessions.csv')
order_details = pd.read_excel(file_path, sheet_name='OrderDetails.csv')

# Display the first few rows of each sheet
user_details_preview = user_details.head()
cooking_sessions_preview = cooking_sessions.head()
order_details_preview = order_details.head()

user_details_preview, cooking_sessions_preview, order_details_preview
```

```python
# Merge cooking sessions with user details first
merged_data = cooking_sessions.merge(user_details, on='User ID', how='left')

# Merge the result with order details using session ID
merged_data = merged_data.merge(order_details, on=['User ID', 'Session ID'], how='left')

# Check the merged dataset structure
merged_preview = merged_data.head()
merged_summary = merged_data.info()

merged_preview, merged_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Analyze relationship between session ratings and order ratings
plt.figure(figsize=(8, 6))
sns.scatterplot(data=merged_data, x='Session Rating', y='Rating', hue='Order Status', style='Order Status')
plt.title('Session Rating vs Order Rating')
plt.xlabel('Session Rating')
plt.ylabel('Order Rating')
plt.legend(title='Order Status')
plt.grid(True)
plt.show()

# Analyze session duration vs order amount
plt.figure(figsize=(8, 6))
sns.scatterplot(data=merged_data, x='Duration (mins)', y='Amount (USD)', hue='Order Status', style='Order Status')
plt.title('Session Duration vs Order Amount')
plt.xlabel('Session Duration (mins)')
plt.ylabel('Order Amount (USD)')
plt.legend(title='Order Status')
plt.grid(True)
plt.show()

```

```python
# Identify popular dishes based on frequency in the order table
popular_dishes = merged_data['Dish Name_y'].value_counts().reset_index()
popular_dishes.columns = ['Dish Name', 'Order Count']

# Display top 5 popular dishes
top_dishes = popular_dishes.head(5)
top_dishes
```

```python
# Plot the popularity of dishes
plt.figure(figsize=(10, 6))
sns.barplot(data=popular_dishes, x='Dish Name', y='Order Count', palette='viridis')
plt.title('Popularity of Dishes')
plt.xlabel('Dish Name')
plt.ylabel('Order Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Analyze demographic factors influencing user behavior

# Group by Age to see average spending and total orders
age_analysis = merged_data.groupby('Age').agg({'Amount (USD)': ['mean', 'sum'], 'Order ID': 'count'})
age_analysis.columns = ['Avg Spend (USD)', 'Total Spend (USD)', 'Order Count']
age_analysis = age_analysis.sort_index()

# Group by Location to check total orders and spending
location_analysis = merged_data.groupby('Location').agg({'Amount (USD)': 'sum', 'Order ID': 'count'})
location_analysis.columns = ['Total Spend (USD)', 'Order Count']
location_analysis = location_analysis.sort_values(by='Order Count', ascending=False)

# Group by Favorite Meal to see preferences
meal_analysis = merged_data.groupby('Favorite Meal').agg({'Amount (USD)': 'sum', 'Order ID': 'count'})
meal_analysis.columns = ['Total Spend (USD)', 'Order Count']
meal_analysis = meal_analysis.sort_values(by='Order Count', ascending=False)

age_analysis, location_analysis, meal_analysis
```

```python
# Visualizing demographic analysis

# Age Analysis - Barplot for order count by age
plt.figure(figsize=(12, 5))
sns.barplot(x=age_analysis.index, y=age_analysis['Order Count'], palette='muted')
plt.title('Orders by Age Group')
plt.xlabel('Age')
plt.ylabel('Order Count')
plt.grid(True)
plt.show()

# Location Analysis - Barplot for total orders by location
plt.figure(figsize=(12, 5))
sns.barplot(x=location_analysis.index, y=location_analysis['Order Count'], palette='viridis')
plt.title('Orders by Location')
plt.xlabel('Location')
plt.ylabel('Order Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Favorite Meal Analysis - Pie chart for meal preferences
plt.figure(figsize=(8, 8))
plt.pie(meal_analysis['Order Count'], labels=meal_analysis.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Meal Preferences')
plt.axis('equal')
plt.show()

```
